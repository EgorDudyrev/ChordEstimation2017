# Описание ноутбука
Содержит скрипты, переводящие аудиофайлы из формата mp3 в wav. Это необходимо, т.к. на момент их написания ещё не существовало библиотек для работы с mp3 файлами

# Пример для одного файла

```python
from pydub import AudioSegment
sound = AudioSegment.from_mp3("audio/billboard/1244/The_Beatles-Help!.mp3")
mono = sound.set_channels(1)
mono.export("audio/billboard/1244/audio.wav", format="wav")
```

# Для всех файлов

```python
import pandas as pd
from pydub import AudioSegment
```

```python
songs_ds = pd.read_csv('McGill-Billboard (2)/songs_downloaded.csv',index_col=0)
songs_ds['converted'] = None
songs_downloaded = songs_ds[songs_ds['status']=='Downloaded']
songs_downloaded.head()
```

```python
def format_fname(s, space_replacer='_'):
    for c in ['(',')','\'']:
        s = s.replace(c,'')
    s = s.strip(' ')
    s = s.replace(' ', space_replacer)
    return s
```

```python
%%time

for song in songs_downloaded.iterrows():
    idx = song[0]
    song = song[1]
    
    print idx,
    dname = 'audio/billboard/'+'0'*(4-len(str(song.number)))+str(song.number)+'/'
    fname = format_fname('{}-{}.mp3'.format(song.artist,song.title))
    try:
        sound = AudioSegment.from_mp3(dname+fname)
        mono = sound.set_channels(1)
        mono.export(dname+"audio.wav", format="wav")
        songs_ds.at[idx, 'converted'] = True
        print '+'
    except:
        songs_ds.at[idx, 'converted'] = False
        print '-'
```

```python
songs_ds.to_csv('McGill-Billboard (2)/songs_converted.csv')
```
